```python

class Nodo:
    def __init__(self, all: list) -> None:
        if len(all) == 14:
            self.title = all[0]
            self.department = all[1]
            self.city = all[2]
            self.property_type = all[3]
            self.latitude = all[4]
            self.longitude = all[5]
            self.surface_total = all[6]
            self.surface_covered = all[7]
            self.bedrooms = all[8]
            self.bathrooms = all[9]
            self.operation_type = all[10]
            self.price = all[11]
            self.mtrc = all[12]
            self.mtrc2 = all[13]
            self.left = None
            self.right = None
class AVL:
    def valores(self, node):
      # Recibe un nodo 'node' como argumento y extrae sus atributos.
      if node is not None:
        # Si el nodo no es None (existe), extrae sus atributos y los devuelve en una lista.
        return [node.title, node.department, node.city, node.property_type,
                node.latitude,
                node.longitude,
                node.surface_total,
                node.surface_covered,
                node.bedrooms,
                node.bathrooms,
                node.operation_type,
                node.price,
                node.mtrc,
                node.mtrc2]
      else:
        return ['este man es un None']
    def insert(self, all):
        self.raiz = self._insert(self.raiz, all)
    def iss(self, node) -> bool:
        act = []
        act = buildpostTree(self.postorder(), self.inorder(), act)
        if self.raiz is not None:
          for i in range(len(act)):
              if self.iguales(node, Nodo(act[i])):
                  return True
          return False

    def _insert(self, node, all):
        if node is None:
            # Cuando el nodo actual es None, se crea un nuevo nodo con los datos de la lista 'all'.
            return Nodo(all)
        elif all[12] < node.mtrc:
            node.left = self._insert(node.left, all)
        elif all[12] > node.mtrc:
            node.right = self._insert(node.right, all)
        elif all[12] == node.mtrc:
            # Cuando 'all[12]' es igual a la métrica 'mtrc' del nodo actual, se compara 'all[13]' con 'node.mtrc2'.
            if all[13] < node.mtrc2:
                node.left = self._insert(node.left, all)
            elif all[13] > node.mtrc2:
                # En caso de que 'all[13]' sea mayor que 'node.mtrc2', se intenta insertar en el subárbol derecho.
                node.right = self._insert(node.right, all)

        # Después de la inserción, se realiza un reequilibrio del árbol llamando a la función 'balance'.
        return self.balance(node)

    def _inorder_int(self, node, conjunto) -> None:
        if node is not None:
            self._inorder_int(node.left, conjunto)
            conjunto.append(self.valores(node))
            self._inorder_int(node.right, conjunto)

    def inorder(self) -> list:
        conjunto = []
        self._inorder_int(self.raiz, conjunto)
        return conjunto

    def _postorder_int(self, node, conjunto):
        if node is not None:
            self._postorder_int(node.left, conjunto)
            self._postorder_int(node.right, conjunto)
            conjunto.append(self.valores(node))

    def postorder(self):
        conjunto = []
        self._postorder_int(self.raiz, conjunto)
        return conjunto

    def __init__(self) -> None:
        self.raiz = None

    def add(self, all, node=None)->None:
      new_node = Nodo(all)

      if self.raiz is None:
          self.raiz = new_node
      else:
        if self.iss(new_node) is not True:
          if node is None:
              node = self.raiz

          if new_node.mtrc < node.mtrc:
              if node.left is None:
                  node.left = new_node

              else:
                  self.add(all, node=node.left)
          elif new_node.mtrc > node.mtrc:
              if node.right is None:
                  node.right = new_node

              else:
                  self.add(all, node=node.right)
          elif new_node.mtrc==node.mtrc:
            #USO DE LA SEGUNDA METRICA
            ##USO DE LA SEGUNDA METRICA
            if new_node.mtrc2 < node.mtrc2:
              if node.left is None:
                  node.left = new_node

              else:
                  self.add(all, node=node.left)
            elif new_node.mtrc2 > node.mtrc2:
              if node.right is None:
                  node.right = new_node

              else:
                  self.add(all, node=node.right)

    def search_by_metric(self, metric, metric2=None):
      # Recibe la métrica principal 'metric' y la métrica opcional 'metric2'
      return self._search_by_metric(self.raiz, metric, metric2)


    def _search_by_metric(self, node, metric, metric2):
    # Recibe un nodo actual 'node' y las métricas 'metric' y 'metric2' a buscar

        if node is None:
            # Si el nodo actual es None, no se encontró el nodo, se retorna None
            return None

        if metric < node.mtrc:
            # Si 'metric' es menor que la métrica del nodo actual, busca en el subárbol izquierdo
            return self._search_by_metric(node.left, metric, metric2)
        elif metric > node.mtrc:
            # Si 'metric' es mayor que la métrica del nodo actual, busca en el subárbol derecho
            return self._search_by_metric(node.right, metric, metric2)
        elif metric == node.mtrc:
            if metric2 is None or metric2 == node.mtrc2:
                # Si 'metric2' es None o igual a la segunda métrica del nodo actual,
                return node
            elif metric2 < node.mtrc2:
                # busca en el subárbol izquierdo
                return self._search_by_metric(node.left, metric, metric2)
            else:
                # Si 'metric2' es mayor que la segunda métrica del nodo actual,
                # busca en el subárbol derecho
                return self._search_by_metric(node.right, metric, metric2)

        return None


    def convert_to_avl(self):
        # Método para convertir un árbol en una estructura AVL
        orden = []

        # Llama a la función 'buildpostTree' para construir el árbol AVL
        buildpostTree(self.postorder(), self.inorder(), orden)
        self.raiz = None

        # Insertar los nodos en el nuevo orden en el árbol actual
        for node in orden:
            self.insert(node)

def search_by_metric(self, metric, metric2=None):
    # Llama al método search by metric con la raíz del árbol y las métricas dadas.
    return self._search_by_metric(self.raiz, metric, metric2)


    def _search_by_metric(self, node, metric, metric2):
      if node is None:
          # Si el nodo actual es None (fin del subárbol), no se encontró el nodo, se retorna None.
          return None

      if metric < node.mtrc:
          return self._search_by_metric(node.left, metric, metric2)
      elif metric > node.mtrc:
          # Si 'metric' es mayor que la métrica del nodo actual, busca en el subárbol derecho.
          return self._search_by_metric(node.right, metric, metric2)
      elif metric == node.mtrc:
          # Si 'metric' es igual a la métrica del nodo actual.
          if metric2 is None or metric2 == node.mtrc2:
              return node
          elif metric2 < node.mtrc2:
              # Si 'metric2' es menor que la segunda métrica del nodo actual,
              # busca en el subárbol izquierdo.
              return self._search_by_metric(node.left, metric, metric2)
          else:
              return self._search_by_metric(node.right, metric, metric2)

      # Si no se encontró el nodo en ninguna de las condiciones anteriores, retorna None.
      return None

    def find_node(self, all):
      return self._find_node(self.raiz, all)

def _find_node(self, node, all):

    if node is None:
        return None

    if all[12] < node.mtrc:
        # Si el valor en la posición 12 de la lista 'all' es menor que la métrica 'mtrc' del nodo actual,
        # busca en el subárbol izquierdo.
        return self._find_node(node.left, all)
    elif all[12] > node.mtrc:
        return self._find_node(node.right, all)
    elif all[12] == node.mtrc and all[13] != node.mtrc2:
        # Si el valor en la posición 12 de la lista 'all' es igual a la métrica 'mtrc' del nodo actual,
        # pero el valor en la posición 13 no coincide con la métrica 'mtrc2', continúa la búsqueda.
        if all[13] < node.mtrc2:
            return self._find_node(node.left, all)
        elif all[13] > node.mtrc2:
            # Si el valor en la posición 13 de la lista 'all' es mayor que la métrica 'mtrc2' del nodo actual,
            # busca en el subárbol derecho.
            return self._find_node(node.right, all)
    return node

    #######################################################


    def iguales(self, node1, node2) -> bool:
    # Recibe dos nodos 'node1' y 'node2' como argumentos y devuelve True si todos sus atributos son iguales, False en caso contrario.

      return (node1.title == node2.title and
                node1.department == node2.department and
                node1.city == node2.city and
                node1.property_type == node2.property_type and
                node1.latitude == node2.latitude and
                node1.longitude == node2.longitude and
                node1.surface_total == node2.surface_total and
                node1.surface_covered == node2.surface_covered and
                node1.bedrooms == node2.bedrooms and
                node1.bathrooms == node2.bathrooms and
                node1.operation_type  == node2.operation_type and
                node1.price == node2.price and
                node1.mtrc == node2.mtrc and
                node1.mtrc2 == node2.mtrc2)


    #######################################################

    def height(self, node):
    # Método para calcular la altura de un nodo en el árbol AVL
        if node is None:
            # Si el nodo es None (es decir, no existe), su altura es -1.
            return -1
        else:
            return max(self.height(node.left), self.height(node.right)) + 1


    #######################################################
    def rotate_right(self, y):
    # Paso 1: Se almacena el hijo izquierdo de 'y' en una variable 'x'.
        x = y.left
        # Paso 2: El hijo derecho de 'x' se convierte en el nuevo hijo izquierdo de 'y'.
        y.left = x.right
        # Paso 3: 'y' se convierte en el nuevo hijo derecho de 'x'.
        x.right = y
        # Paso 4: Se devuelve 'x', que ahora es el nuevo nodo raíz del subárbol rotado.
        return x


    #######################################################
    def rotate_left(self, x):
        y = x.right
        x.right = y.left
        y.left = x
        return y

    #######################################################
    # def iss(self, node) -> bool:
    #     act = []
    #     act = buildpostTree(self.postorder(), self.inorder(), act)
    #     if self.raiz is not None:
    #       for i in range(len(act)):
    #           if self.iguales(node, Nodo(act[i])):
    #               return True
    #       return False


    #######################################################
    #######################################################
    def insert(self, all):
        self.raiz = self._insert(self.raiz, all)

    #######################################################
    def balance(self, node):
        if node is None:
            return node
        if self.height(node.left) - self.height(node.right) > 1:
            if self.height(node.left.left) >= self.height(node.left.right):
                node = self.rotate_right(node)
            else:
                node.left = self.rotate_left(node.left)
                node = self.rotate_right(node)

        elif self.height(node.right) - self.height(node.left) > 1:
            if self.height(node.right.right) >= self.height(node.right.left):
                node = self.rotate_left(node)
            else:
                node.right = self.rotate_right(node.right)
                node = self.rotate_left(node)

        return node


    #######################################################
    def arbol_actual(self) -> List:
        orden = []
        return buildpostTree(self.postorder(), self.inorder(), orden)

    #######################################################
    def delete(self, all) -> Any:
          self.raiz = self._delete(self.raiz, all)

    def _delete(self, node, all):
        if node is None:
            # no se encuentra el nodo a eliminar, así que se retorna el nodo actual.
            return node

        if all[12] < node.mtrc:
            # se busca el nodo a eliminar en el subárbol izquierdo.
            node.left = self._delete(node.left, all)
        elif all[12] > node.mtrc:
            node.right = self._delete(node.right, all)
        else:
            if all[12] == node.mtrc:
                # Cuando 'all[12]' es igual a la métrica 'mtrc' del nodo actual, se compara 'all[13]' con 'node.mtrc2'.
                if all[13] < node.mtrc2:
                    node.left = self._delete(node.left, all)
                elif all[13] > node.mtrc2:
                    node.right = self._delete(node.right, all)
                else:
                    # Cuando se encuentra el nodo a eliminar.
                    if node.left is None:
                        # Si el nodo a eliminar no tiene hijo izquierdo, se reemplaza por su hijo derecho (o None si no tiene).
                        temp = node.right
                        node = None
                        return temp
                    elif node.right is None:
                        # Si el nodo a eliminar no tiene hijo derecho, se reemplaza por su hijo izquierdo.
                        temp = node.left
                        node = None
                        return temp
                    # Si el nodo a eliminar tiene ambos hijos, se reemplaza por su sucesor en orden.
                    temp = self.sucesor(node.right)
                    node.mtrc, node.mtrc2 = temp.mtrc, temp.mtrc2
                    node.right = self._delete(node.right, all)
        return self.balance(node)

    #######################################################
    def sucesor(self, node)-> Any:
        current = node
        while current.left is not None:
            current = current.left
        return current
    #######################################################
    def buscar_nivel(self, all, nvl=0, node=None) -> int:
        Wanted = Nodo(all)
        if node is None:
            # Si el nodo actual es None, no se encuentra el nodo buscado, por lo que se retorna -1.
            return -1
        if self.iss(Wanted):
            if self.iguales(Wanted, self.raiz):
                # Si el nodo buscado es igual a la raíz del árbol, se indica que están en la misma raíz.
                print("el nodo esta en la misma raíz")
                return nvl

            if Wanted.mtrc == node.mtrc and Wanted.mtrc2 == node.mtrc2:
                # Si los datos del nodo buscado son iguales a los datos del nodo actual, se retorna el nivel actual.
                return nvl
            if Wanted.mtrc < node.mtrc:
                return self.buscar_nivel(all, nvl=nvl + 1, node=node.left)
            elif Wanted.mtrc > node.mtrc:
                return self.buscar_nivel(all, nvl=nvl + 1, node=node.right)
            elif Wanted.mtrc == node.mtrc and Wanted.mtrc2 != node.mtrc2:
                # Si Wanted.mtrc es igual a node.mtrc pero Wanted.mtrc2 no es igual a node.mtrc2,
                if Wanted.mtrc2 < node.mtrc2:
                    return self.buscar_nivel(all, nvl=nvl + 1, node=node.left)
                elif Wanted.mtrc2 > node.mtrc2:
                    return self.buscar_nivel(all, nvl=nvl + 1, node=node.right)
        # Si no se encuentra el nodo buscado en el árbol, se retorna -1.
        return -1

    def balance_f(self,node):
      if node is not None:
        return self.height(node.right)-self.height(node.left)
    def padre(self, node, all):
        # una lista 'all' que contiene los datos del nodo hijo y busca su padre.
        if node is None:
            # Si el nodo actual es None, no se encuentra el nodo hijo en el árbol, por lo que no hay padre.
            return None
        if all[12] < node.mtrc:
            if self.iguales(node.left, Nodo(all)):
                # Si el nodo hijo es igual al nodo izquierdo del nodo actual, se devuelve la información del nodo padre.
                return self.valores(node)
            return self.padre(node.left, all)
        elif all[12] > node.mtrc:
            if self.iguales(node.right, Nodo(all)):
                return self.valores(node)
            return self.padre(node.right, all)
        elif all[12] == node.mtrc and all[13] != node.mtrc2:
            if all[13] < node.mtrc2:
                # Si 'all[13]' es menor que 'node.mtrc2', se verifica el subárbol izquierdo.
                if self.iguales(node.left, Nodo(all)):
                    # Si el nodo hijo es igual al nodo izquierdo del nodo actual, se devuelve la información del nodo padre.
                    return self.valores(node)
                return self.padre(node.left, all)
            elif all[13] > node.mtrc2:
                if self.iguales(node.right, Nodo(all)):
                    return self.valores(node)
                return self.padre(node.right, all)

        # Si no se encuentra el nodo hijo en el árbol, se retorna None para indicar que no tiene padre.
        return None

    def abuelo(self,hijito):
      padre=self.padre(self.raiz,hijito)
      if padre is not None:
        abuelo=self.padre(self.raiz,padre)
        if abuelo is not None:
          print("hola, soy el abuelito :>")
          return abuelo
        else:
          print("no hay abuelo, ya nada")
      else:
        return "no ni si quiera padre"
    def tio(self, node1, node2, node3, elem):
        # Método para encontrar el tío de un nodo con datos cel arbol AVL.
        if node1 is not None:
            if (elem[12] == node1.mtrc) or (elem[13] == node1.mtrc2):
                return node3
            elif elem[12] < node1.mtrc:
                if node2 is not None:
                    if node1 is node2.left:
                        # se busca el tío en el subárbol derecho del nodo2 y se establece el nodo1 como el nuevo nodo2.
                        return self.tio(node1.left, node1, node2.right, elem)
                    else:
                        # Si el nodo1 es el hijo derecho del nodo2 (padre del nodo1),
                        return self.tio(node1.left, node1, node2.left, elem)
                # Si no hay nodo2 (padre), se busca el tío en el subárbol izquierdo del nodo1 y se establece nodo2 como None.
                return self.tio(node1.left, node1, None, elem)
            elif elem[13] < node1.mtrc2:
                if node2 is not None:
                    if node1 is node2.left:
                        return self.tio(node1.left, node1, node2.right, elem)
                    else:
                        return self.tio(node1.left, node1, node2.left, elem)
                return self.tio(node1.left, node1, None, elem)
            else:
                if node2 is not None:
                    if node1 is node2.left:
                        # Si el nodo1 es el hijo izquierdo del nodo2
                        return self.tio(node1.right, node1, node2.right, elem)
                    else:
                        return self.tio(node1.right, node1, node2.left, elem)
                return self.tio(node1.right, node1, None, elem)
        # Si el nodo1 es None, no se encuentra el nodo de referencia en el árbol, por lo que no hay tío.
        return None


    #BELLEZA, HERMOSURA, TIFON
    def inlevel(self, nodes, nvls, max, n, out=[], j=0, i=0, node=None):
        # el nivel max hasta el cual se imprimirán nodos, la cantidad total de nodos, una lista 'out' para llevar registro de nodos ya impresos,
        if j < n and nvls[i] < max:
            node = nodes[j]
            if (self.buscar_nivel(node, 0, self.raiz) == nvls[i]) and not (node in out):
                # Si el nivel del nodo actual coincide con el nivel en 'nvls' y el nodo no ha sido impreso previamente:
                print(f"Nivel: {self.buscar_nivel(node, 0, self.raiz)}")
                print(node)
                out = nodes.pop(j)  # Se remueve el nodo de 'nodes' y se agrega a 'out' para evitar impresiones duplicadas.
                self.inlevel(nodes, nvls, max, n, out=out, j=0, i=i + 1)
            else:
                # Si el nodo actual no coincide con el nivel en 'nvls' o ya ha sido impreso, se procesa el siguiente nodo en 'nodes'.
                self.inlevel(nodes, nvls, max, n, out=out, j=j + 1, i=i)
        elif j < len(nodes) and nvls[i] == max:
            node = nodes[j]
            if (self.buscar_nivel(node, 0, self.raiz) == nvls[i]) and not (node in out):
                print(f"Nivel: {self.buscar_nivel(node, 0, self.raiz)}")
                print(node)
                out = nodes.pop(j)
                self.inlevel(nodes, nvls, max, n, out=out, j=0, i=i)
            else:
                self.inlevel(nodes, nvls, max, n, out=out, j=j + 1, i=i)
        elif j == len(nodes) and nvls[i] == max:
            # Cuando se ha procesado todos los nodos en el nivel actual, se imprime el nivel y el último nodo.
            print(f"Nivel: {nvls[i]}")
            print(node)
            return


def mostrar_arbol_avl(raiz, nombre_archivo):
    def agregar_nodo(grafico, nodo):
        if nodo:
            grafico.node(str(nodo.mtrc) )
            if nodo.left:

                grafico.edge(str(nodo.mtrc), str(nodo.left.mtrc))
                agregar_nodo(grafico, nodo.left)
            if nodo.right:
                grafico.edge(str(nodo.mtrc), str(nodo.right.mtrc))
                agregar_nodo(grafico, nodo.right)

    grafo = graphviz.Digraph(format='png')
    agregar_nodo(grafo, raiz)
    grafo.render(nombre_archivo, view=False)

def buildpostTree(postorder: list, inorder: list, Final_tree: list) -> list:
    # Función para construir un árbol a partir de recorridos en postorden e inorden
    # y 'Final_tree' para registrar los valores de los nodos en el recorrido en postorden.
    if not postorder or not inorder:
        return None
    # Se obtiene el valor de la raíz del subárbol a partir del último valor en el recorrido en postorden.
    raiz_val = postorder[-1]
    # Se agrega el valor de la raíz a la lista 'Final_tree', que registra el recorrido en postorden.
    Final_tree.append(raiz_val)
    raiz = Nodo(raiz_val)
    # Se encuentra la posición de la raíz en el recorrido en inorden.
    raiz_index = inorder.index(raiz_val)
    raiz.left = buildpostTree(postorder[:raiz_index], inorder[:raiz_index], Final_tree)
    raiz.right = buildpostTree(postorder[raiz_index:-1], inorder[raiz_index + 1:], Final_tree)
    return raiz
    return Final_tree

def ingreso_datos(key):
    print(key-1)
    tree.add(datoss[key - 1])
    tree.convert_to_avl()
    graficar()
def borrar(key):
    print(key-1)
    print(tree.delete(datoss[key-1]))
def nivel_nodo(key):
    nv=0
    return tree.buscar_nivel(datoss[key-1],nv,tree.raiz)
def balance_Factor(key):
    return (tree.balance_f(tree.find_node(datoss[key-1])))
def father(key):
    print(tree.padre(tree.raiz,datoss[key-1]))
def grand_father(key):
    print(tree.abuelo(datoss[key-1]))
def uncle(key):
    tio=tree.tio(tree.raiz,None,None,datoss[key-1])
    print(tree.valores(tio))

def recorrido_por_niveles(nvls,m):
    inord=tree.inorder()
    tree.inlevel(inord,nvls,m,len(inord))

def location(key) -> tuple:
    # Función para obtener la ubicación de un nodo y generar un mapa.
    if tree.iss(Nodo(datoss[key - 1])):
        # Se obtienen las coordenadas de latitud y longitud, así como la ciudad del nodo.
        lat, longt, city = datoss[key - 1][4], datoss[key - 1][5], datoss[key - 1][2]
        # Imprime un mensaje con la ciudad correspondiente.
        print("Esta es la localización de:", city)
        # Se crea un mapa con la biblioteca 'folium' utilizando las coordenadas de latitud y longitud.
        mapa = fl.Map(location=[lat, longt])
        # Se guarda el mapa generado como un archivo HTML llamado "mapa_base.html".
        mapa.save("mapa_base.html")
        # Retorna una tupla con las coordenadas de latitud y longitud.
        return lat, longt


def graficar():
  name = "avl"
  mostrar_arbol_avl(tree.raiz, name)
  # Abrir la imagen generada
  img = Image.open(name+ ".png")
  img.show()
  return IPImage(name + ".png")
```

```python
class Nodo:
    def __init__(self, all: list) -> None:
        if len(all) == 14:
            self.title = all[0]
            self.department = all[1]
            self.city = all[2]
            self.property_type = all[3]
            self.latitude = all[4]
            self.longitude = all[5]
            self.surface_total = all[6]
            self.surface_covered = all[7]
            self.bedrooms = all[8]
            self.bathrooms = all[9]
            self.operation_type = all[10]
            self.price = all[11]
            self.mtrc = all[12]
            self.mtrc2 = all[13]
            self.left = None
            self.right = None
class AVL:

    def __init__(self) -> None:
        self.raiz = None

    def search_by_criteria(self, criteria_values):
      results = []
      self._search_by_criteria(self.raiz, criteria_values, results)
      return results

    def _search_by_criteria(self, node, criteria_values, results):
        if node is None:
            return

        match = True

        for i, value in enumerate(criteria_values):
            if i == 0 and float(value) != node.mtrc:
                match = False
                break
            elif i == 1 and float(value) != node.mtrc2:
                match = False
                break
            elif i == 2 and value != node.title:
                match = False
                break
            elif i == 3 and value != node.city:
                match = False
                break
            elif i == 4 and value != node.property_type:
                match = False
                break
            elif i == 5 and float(value) != node.latitude:
                match = False
                break
            elif i == 6 and float(value) != node.longitude:
                match = False
                break
            elif i == 7 and float(value) != node.surface_total:
                match = False
                break
            elif i == 8 and float(value) != node.surface_covered:
                match = False
                break
            elif i == 9 and int(value) != node.bedrooms:
                match = False
                break
            elif i == 10 and int(value) != node.bathrooms:
                match = False
                break
            elif i == 11 and value != node.operation_type:
                match = False
                break
            elif i == 12 and float(value) != node.price:
                match = False
                break
            elif i == 13 and float(value) != node.mtrc:
                match = False
                break
            elif i == 14 and float(value) != node.mtrc2:
                match = False
                break

        if match:
            results.append(self.valores(node))

        self._search_by_criteria(node.left, criteria_values, results)
        self._search_by_criteria(node.right, criteria_values, results)

    def add(self, all, node=None)->None:
      new_node = Nodo(all)

      if self.raiz is None:
          self.raiz = new_node
      else:
        if self.iss(new_node) is not True:
          if node is None:
              node = self.raiz

          if new_node.mtrc < node.mtrc:
              if node.left is None:
                  node.left = new_node

              else:
                  self.add(all, node=node.left)
          elif new_node.mtrc > node.mtrc:
              if node.right is None:
                  node.right = new_node

              else:
                  self.add(all, node=node.right)
          elif new_node.mtrc==node.mtrc:
            #USO DE LA SEGUNDA METRICA Y FALTA EL DELETE
            ##USO DE LA SEGUNDA METRICA
            if new_node.mtrc2 < node.mtrc2:
              if node.left is None:
                  node.left = new_node

              else:
                  self.add(all, node=node.left)
            elif new_node.mtrc2 > node.mtrc2:
              if node.right is None:
                  node.right = new_node

              else:
                  self.add(all, node=node.right)

    def search_by_metric(self, metric, metric2=None):
      # Recibe la métrica principal 'metric' y la métrica opcional 'metric2'
      return self._search_by_metric(self.raiz, metric, metric2)


    def _search_by_metric(self, node, metric, metric2):
    # Recibe un nodo actual 'node' y las métricas 'metric' y 'metric2' a buscar

        if node is None:
            # Si el nodo actual es None, no se encontró el nodo, se retorna None
            return None

        if metric < node.mtrc:
            # Si 'metric' es menor que la métrica del nodo actual, busca en el subárbol izquierdo
            return self._search_by_metric(node.left, metric, metric2)
        elif metric > node.mtrc:
            # Si 'metric' es mayor que la métrica del nodo actual, busca en el subárbol derecho
            return self._search_by_metric(node.right, metric, metric2)
        elif metric == node.mtrc:
            if metric2 is None or metric2 == node.mtrc2:
                # Si 'metric2' es None o igual a la segunda métrica del nodo actual,
                return node
            elif metric2 < node.mtrc2:
                # busca en el subárbol izquierdo
                return self._search_by_metric(node.left, metric, metric2)
            else:
                # Si 'metric2' es mayor que la segunda métrica del nodo actual,
                # busca en el subárbol derecho
                return self._search_by_metric(node.right, metric, metric2)

        return None

    def convert_to_avl(self):
      orden = []
      buildpostTree(self.postorder(), self.inorder(), orden)
      self.raiz = None
      for node in orden:
        self.insert(node)
    def find_node(self, all):
        return self._find_node(self.raiz, all)

    def _find_node(self, node, all):
        if node is None:
            return None
        if all[12] < node.mtrc:
            return self._find_node(node.left, all)
        elif all[12] > node.mtrc:
            return self._find_node(node.right, all)
        elif all[12] == node.mtrc and all[13]!=node.mtrc2:
            if all[13] < node.mtrc2:
                return self._find_node(node.left, all)
            elif all[13] > node.mtrc2:
                return self._find_node(node.right, all)
        return node


    #######################################################
    def valores(self, node):
      if node is not None:
        return [node.title, node.department, node.city, node.property_type,
                node.latitude,
                node.longitude,
                node.surface_total,
                node.surface_covered,
                node.bedrooms,
                node.bathrooms,
                node.operation_type,
                node.price,
                node.mtrc,
                node.mtrc2]
      else:
        return ['este man es un None']
    def _inorder_int(self, node, conjunto) -> None:
        if node is not None:
            self._inorder_int(node.left, conjunto)
            conjunto.append(self.valores(node))
            self._inorder_int(node.right, conjunto)

    def inorder(self) -> list:
        conjunto = []
        self._inorder_int(self.raiz, conjunto)
        return conjunto

    def _postorder_int(self, node, conjunto):
        if node is not None:
            self._postorder_int(node.left, conjunto)
            self._postorder_int(node.right, conjunto)
            conjunto.append(self.valores(node))

    def postorder(self):
        conjunto = []
        self._postorder_int(self.raiz, conjunto)
        return conjunto
    def iguales(self, node1, node2) -> bool:
        return (node1.title == node2.title and
                node1.department == node2.department and
                node1.city == node2.city and
                node1.property_type == node2.property_type and
                node1.latitude == node2.latitude and
                node1.longitude == node2.longitude and
                node1.surface_total == node2.surface_total and
                node1.surface_covered == node2.surface_covered and
                node1.bedrooms == node2.bedrooms and
                node1.bathrooms == node2.bathrooms and
                node1.operation_type == node2.operation_type and
                node1.price == node2.price and
                node1.mtrc == node2.mtrc and
                node2.mtrc2 == node2.mtrc2)

    #######################################################

    def height(self, node):
        if node is None:
            return -1
        return max(self.height(node.left), self.height(node.right)) + 1

    #######################################################
    def rotate_right(self, y):
        x = y.left
        y.left = x.right
        x.right = y
        return x

    #######################################################
    def rotate_left(self, x):
        y = x.right
        x.right = y.left
        y.left = x
        return y

    #######################################################
    def iss(self, node) -> bool:
        act = []
        act = buildpostTree(self.postorder(), self.inorder(), act)
        if self.raiz is not None:
          for i in range(len(act)):
              if self.iguales(node, Nodo(act[i])):
                  return True
          return False

    #######################################################
    #######################################################
    def insert(self, all):

        self.raiz = self._insert(self.raiz, all)

    #######################################################
    def balance(self, node):
        if node is None:
            return node
        if self.height(node.left) - self.height(node.right) > 1:
            if self.height(node.left.left) >= self.height(node.left.right):
                node = self.rotate_right(node)
            else:
                node.left = self.rotate_left(node.left)
                node = self.rotate_right(node)

        elif self.height(node.right) - self.height(node.left) > 1:
            if self.height(node.right.right) >= self.height(node.right.left):
                node = self.rotate_left(node)
            else:
                node.right = self.rotate_right(node.right)
                node = self.rotate_left(node)

        return node

    #######################################################

    #######################################################
    def _insert(self, node, all):
      if node is None:
          return Nodo(all)
      if all[12] < node.mtrc:
          node.left = self._insert(node.left, all)
      elif all[12] > node.mtrc:
          node.right = self._insert(node.right, all)
      elif all[12]==node.mtrc:
        if all[13] < node.mtrc2:
            node.left = self._insert(node.left, all)
        elif all[13] > node.mtrc2:
            node.right = self._insert(node.right, all)
      return self.balance(node)


    #######################################################
    def arbol_actual(self) -> List:
        orden = []
        return buildpostTree(self.postorder(), self.inorder(), orden)

    #######################################################
    def delete(self, all) -> Any:
          self.raiz = self._delete(self.raiz, all)

    def _delete(self, node, all):
      if node is None:
          return node

      if all[12] < node.mtrc:
          node.left = self._delete(node.left, all)
      elif all[12] > node.mtrc:
          node.right = self._delete(node.right, all)
      else:
          if all[12] == node.mtrc:
              if all[13] < node.mtrc2:
                  node.left = self._delete(node.left, all)
              elif all[13] > node.mtrc2:
                  node.right = self._delete(node.right, all)
              else:
                  if node.left is None:
                      temp = node.right
                      node = None
                      return temp
                  elif node.right is None:
                      temp = node.left
                      node = None
                      return temp
                  temp = self.sucesor(node.right)
                  node.mtrc, node.mtrc2 = temp.mtrc, temp.mtrc2
                  node.right = self._delete(node.right, all)
      return self.balance(node)

    #######################################################
    def sucesor(self, node)-> Any:
        current = node
        while current.left is not None:
            current = current.left
        return current
    #######################################################
    def buscar_nivel(self, all, nvl=0, node=None) -> int:
      Wanted = Nodo(all)
      if node is None:
          return -1
      if self.iss(Wanted):

          if self.iguales(Wanted, self.raiz):
              print("misma raíz")
              return nvl
          if Wanted.mtrc == node.mtrc and Wanted.mtrc2 == node.mtrc2:

              return nvl
          if Wanted.mtrc < node.mtrc:
              return self.buscar_nivel(all, nvl=nvl + 1, node=node.left)
          elif Wanted.mtrc > node.mtrc:
              return self.buscar_nivel(all, nvl=nvl + 1, node=node.right)
          elif Wanted.mtrc == node.mtrc and Wanted.mtrc2 != node.mtrc2:
              if Wanted.mtrc2 < node.mtrc2:
                  return self.buscar_nivel(all, nvl=nvl + 1, node=node.left)
              elif Wanted.mtrc2 > node.mtrc2:
                  return self.buscar_nivel(all, nvl=nvl + 1, node=node.right)
      return -1
    def balance_f(self,node):
      if node is not None:
        return self.height(node.right)-self.height(node.left)
    def padre(self,node, all):
        if node is None:
            return None
        if all[12] < node.mtrc:
            if self.iguales(node.left, Nodo(all)):
                return self.valores(node)
            return self.padre(node.left, all)
        elif all[12] > node.mtrc:
            if self.iguales(node.right, Nodo(all)):
                return self.valores(node)
            return self.padre(node.right, all)
        elif all[12] == node.mtrc and all[13] != node.mtrc2:
            if all[13] < node.mtrc2:
                if self.iguales(node.left, Nodo(all)):
                    return self.valores(node)
                return self.padre(node.left, all)
            elif all[13] > node.mtrc2:
                if self.iguales(node.right, Nodo(all)):
                    return self.valores(node)
                return self.padre(node.right, all)
        return None
    def abuelo(self,hijito):
      padre=self.padre(self.raiz,hijito)
      if padre is not None:
        abuelo=self.padre(self.raiz,padre)
        if abuelo is not None:
          print("abuelito")
          return abuelo
        else:
          print("no hay abuelo")
      else:
        return "no ni si quiera padre"
    def tio(self,node1, node2, node3, elem):
      if node1 is not None:
          if (elem[12] == node1.mtrc) or (elem[13] == node1.mtrc2):
              return node3
          elif elem[12] < node1.mtrc:
              if node2 is not None:
                  if node1 is node2.left:
                      return self.tio(node1.left, node1, node2.right, elem)
                  else:
                      return self.tio(node1.left, node1, node2.left, elem)
              return self.tio(node1.left, node1, None, elem)
          elif elem[13] < node1.mtrc2:
              if node2 is not None:
                  if node1 is node2.left:
                      return self.tio(node1.left, node1, node2.right, elem)
                  else:
                      return self.tio(node1.left, node1, node2.left, elem)
              return self.tio(node1.left, node1, None, elem)
          else:
              if node2 is not None:
                  if node1 is node2.left:
                      return self.tio(node1.right, node1, node2.right, elem)
                  else:
                      return self.tio(node1.right, node1, node2.left, elem)
              return self.tio(node1.right, node1, None, elem)
      return None
    #BELLEZA, HERMOSURA
    def inlevel(self,nodes,nvls,max,n,out=[] ,j=0,i=0,node=None):
      if j<n and nvls[i]<max:
          node = nodes[j]
          if (self.buscar_nivel(node, 0, self.raiz) == nvls[i]) and not(node in out) :
              print(f"nivel: {self.buscar_nivel(node, 0, self.raiz)}")
              print(node)
              out=nodes.pop(j)

              self.inlevel(nodes, nvls,max,n,out=out, j=0, i=i + 1)
          else:
              self.inlevel(nodes, nvls,max,n,out=out, j=j + 1, i=i)
      elif j<len(nodes) and nvls[i]==max:
         node = nodes[j]
         if (self.buscar_nivel(node, 0, self.raiz) == nvls[i]) and not(node in out) :
              print(f"nivel: {self.buscar_nivel(node, 0, self.raiz)}")
              print(node)
              out=nodes.pop(j)
              self.inlevel(nodes, nvls,max,n,out=out, j=0, i=i)
         else:
              self.inlevel(nodes, nvls,max,n,out=out, j=j + 1, i=i)
      elif j==len(nodes) and nvls[i]==max:
              print(f"nivel: {nvls[i]}")
              print(node)
              return

def mostrar_arbol_avl(raiz, nombre_archivo):
    def agregar_nodo(grafico, nodo):
        if nodo:
            grafico.node(str(nodo.city) )
            if nodo.left:
                grafico.edge(str(nodo.mtrc), str(nodo.left.mtrc))
                agregar_nodo(grafico, nodo.left)
            if nodo.right:
                grafico.edge(str(nodo.mtrc), str(nodo.right.mtrc))
                agregar_nodo(grafico, nodo.right)

    grafo = graphviz.Digraph(format='png')
    agregar_nodo(grafo, raiz)
    grafo.render(nombre_archivo, view=False)

def buildpostTree(postorder: list, inorder: list, Final_tree: list) -> list:
    if not postorder or not inorder:
        return None

    raiz_val = postorder[-1]
    Final_tree.append(raiz_val)
    raiz = Nodo(raiz_val)

    raiz_index = inorder.index(raiz_val)

    raiz.left = buildpostTree(postorder[:raiz_index], inorder[:raiz_index], Final_tree)
    raiz.right = buildpostTree(postorder[raiz_index:-1], inorder[raiz_index + 1:], Final_tree)

    return Final_tree

def ingreso_datos(key):
    print(key-1)
    tree.add(datoss[key - 1])
    tree.convert_to_avl()
def borrar(key):
    print(key-1)
    print(tree.delete(datoss[key-1]))
def nivel_nodo(key):
    nv=0
    return tree.buscar_nivel(datoss[key-1],nv,tree.raiz)
def balance_Factor(key):
    return (tree.balance_f(tree.find_node(datoss[key-1])))
def father(key):
    print(tree.padre(tree.raiz,datoss[key-1]))
def grand_father(key):
    print(tree.abuelo(datoss[key-1]))
def uncle(key):
    tio=tree.tio(tree.raiz,None,None,datoss[key-1])
    print(tree.valores(tio))

def recorrido_por_niveles(nvls,m):
    inord=tree.inorder()
    tree.inlevel(inord,nvls,m,len(inord))
def location(key)->tuple:
    lat,longt,city=datoss[key-1][4],datoss[key-1][5],datoss[key-1][2]
    print("esta es la localizacion de :",city)
    mapa=fl.Map(location=[lat,longt])
    mapa.save("mapa_base.html")
    mapa

def graficar():
  name = "avl"
  mostrar_arbol_avl(tree.raiz, name)
  # Abrir la imagen generada
  img = Image.open(name+ ".png")
  img.show()
  return IPImage(name + ".png")
```
